```python
import pandas as pd
import openpyxl

file_list = ['2022_年間売上表.xlsx', '2023_年間売上表.xlsx']

df = pd.DataFrame()

for file in file_list:
    temp_df = pd.read_excel(file)
    temp_df.columns = ['商品','売上年','金額']
    df = pd.concat([df, temp_df], ignore_index=True)

print(df)
```

```python
import pandas as pd
import openpyxl
from openpyxl.styles import PatternFill

excel_file_path_2022 = "2022_年間売上表.xlsx"
excel_file_path_2023 = "2023_年間売上表.xlsx"

df_2022 = pd.read_excel(excel_file_path_2022, sheet_name="Sheet1")
df_2023 = pd.read_excel(excel_file_path_2023, sheet_name="Sheet1")

df_combined = pd.concat([df_2022, df_2023])

grouped_df = df_combined.groupby(["商品", "売上年"])["金額（千円）"].sum().reset_index()


grouped_df.to_excel("売上集計表.xlsx", index=False)

workbook = openpyxl.load_workbook("売上集計表.xlsx")
worksheet = workbook.active

header_row = 1

fill = PatternFill(patternType='solid', fgColor='F2F2F2')

for col_num in range(len(grouped_df.columns)):
    cell = worksheet.cell(row=header_row, column=col_num + 1)
    cell.fill = fill

workbook.save("売上集計表.xlsx")
grouped_df
```
